```python
import pandas as pd
import matplotlib.pyplot as plt


file_path = "Salary Data (2).csv"  # Zmień na właściwą ścieżkę pliku
df = pd.read_csv(file_path)

plt.figure(figsize=(10, 6))
plt.hist(df["Salary"], bins=20, color="pink", edgecolor="black", alpha=0.7)

plt.title("Rozkład wynagrodzeń", fontsize=14)
plt.xlabel("Wynagrodzenie", fontsize=12)
plt.ylabel("Liczba osób", fontsize=12)

plt.show()
```
